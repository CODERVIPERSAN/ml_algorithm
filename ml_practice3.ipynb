```python
from sklearn.linear_model import SGDRegressor

sgd_regressor = SGDRegressor(random_state=10)

for i in range(X_train_r.shape[0]):
  sgd_regressor.partial_fit(X_train_r[i],y_train_r[i])
  if i==4:
    print(sgd_regressor.coef_[4])

```
