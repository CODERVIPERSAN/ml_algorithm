```python
import pandas as pd
```

```python


df = pd.read_excel("TownwiseCensus.xls")
name = df['Literates Population Female']/(df['Literates Population Female']+df['Illiterate Female'])
df['female rate']= name
# df =df['female rate'==None]

# Drop rows that has NaN values on selected columns
df=df.dropna(subset=['female rate'])
df[df['female rate']>100]  #verified



```

```python
data ={}
data = df.to_dict()
# print(data)

village = data['Town/Village_Code']

dataprocess={}

literatespop = data['Literates Population Female']
illiterate = data['Illiterate Female']

for i in village:
    # print(i,village[i],literatespop[i])

    # print(village[i] in dataprocess.keys())

    # dataprocess[village[i]]=literatespop[i]
    
    # # print(village[i])
    if(village[i] in dataprocess.keys()):
        # print("hello")
        # print(village[i]in dataprocess.keys())
        dataprocess[village[i]]['literate']+=literatespop[i]
        dataprocess[village[i]]['iliterate']+=illiterate[i]
        dataprocess[village[i]]['Total']+=literatespop[i]+illiterate[i]
    else:
        dataprocess[village[i]]={'literate':literatespop[i],'iliterate':illiterate[i],'Total':literatespop[i]+illiterate[i]}
        # print(dataprocess[village[i]])


# print( [village[i]: for i in village])


frame = pd.DataFrame(dataprocess)
value={}
max_ratio=-1
max_village=-1
for village in dataprocess.keys():
    print({'town':village,'literate':dataprocess[village]['literate'],'illiterate':dataprocess[village]['iliterate'],'total':dataprocess[village]['Total'],'ratio':dataprocess[village]['literate']/dataprocess[village]['Total']})
    ratio = dataprocess[village]['literate']/dataprocess[village]['Total']
    if max_ratio<ratio:
        max_ratio=ratio
        max_village=village
print(max_ratio)
print(max_village)
pd.DataFrame(value)

# print([ literatespop[i] for i in literatespop])
```
